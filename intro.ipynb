##### Matplotlib 

```python
import matplotlib.pyplot as plt 
```

```python
x =  [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)
plt.show()
```

```python
x = ["Mon", "tues", "wed", "thurs", "fri"]
y = [10,7,15,20,17]

plt.plot(x,y)

plt.title("Bakery Sales per Week")

plt.xlabel("Days of the week")
plt.ylabel("sales per Day")

plt.grid()

plt.show()
```
