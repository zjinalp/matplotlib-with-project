```python
import matplotlib.pyplot as plt 
```

```python
months =  [1,2,3,4]
sales = [1000,1500,1200,1800]
```

### plt.plot() 
###### plt.plot(x,y, color="clrname",linestyle="linestyle",linewidth = value,marker = "marker value",label = "label name")

```python
plt.plot(months, sales, 
         color = "orange", 
         linestyle="--",
         linewidth = 2,
         marker = "o",
         label = "2025 Sales Data")

# labels on axis
plt.xlabel("Months")
plt.ylabel("Sales Per Month")

# title
plt.title("Monthly Sales Data Report")

# to show grid
plt.grid(color = "gray", linestyle=":", linewidth = 1)

# legeng - it shows each line representation
plt.legend()
# location of legent loc="upper left" or "lower right"
# plt.legend(loc = "lower right")
plt.legend(loc = "upper left", fontsize = 20)


# axis limit give the ranges to the plot that data is in this range 
plt.xlim(1,4)
plt.ylim(0,2000)


# axis ticks - make numeric ticks meaningful by giving name 
plt.xticks([1,2,3,4],["Jan","Feb","Mar","Apr"])
plt.yticks()

# display the plot
plt.show()
```
