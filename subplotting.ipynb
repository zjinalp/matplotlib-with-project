```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4]
y = [10,20,25,35]

# plt.figure(figsize=(8,5))

plt.subplot(1,2,1) # 1st row, 2nd col, 1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2) # 1st row, 2nd col, 1st subplot
plt.bar(x,y)
plt.title("Bar Chart")


plt.tight_layout()
plt.show()
```

##### How to customize individual subplot

```python
# more professional way

fig, ax = plt.subplots(1,2, figsize = (10,5))
a = [11,12,13,14]
b = [27,20,45,55]

ax[0].plot(a,b, color = "green")
ax[0].set_title("Line Plot")
ax[0].set_xlabel("X Value of line Chart")
ax[0].set_ylabel("Y Value of line Chart")


ax[1].bar(a,b, color = "orange")
ax[1].set_title("Bar Plot")
ax[1].set_xlabel("X Value of Bar Chart")
ax[1].set_ylabel("Y Value of Bar Chart")

# subtitile for charts
fig.suptitle("Comparison btw Line and Bar chart")

# to fit layout in the canvas
# automatic adjustment
#  just use before plt.show()
plt.tight_layout()

# How to use savefig() function
# savefig( "filename.extention, dpi =  value, bbox_inches = 'tight'")
# bbox_inches means it tightens the outer white space.
plt.savefig("subplots.png", dpi = 300, bbox_inches = "tight")

plt.show()
```
