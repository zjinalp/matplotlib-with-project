```python
import matplotlib.pyplot as plt 
```

#### Bar charts
###### Category comparison Analysis
#### Pie charts
###### Proportional illustration whole representation
#### Histogram
###### Numerical Distribution Data Insights


```python
# bar chart --> plt.bar(x, height, color = "cname", width = value, label = "label name")

product = ["a","b","c","d"]
sales = [1000,1500,800,1200]

plt.bar(product, sales, color ="blue", label = "Sales 2025")

plt.xlabel("Products")
plt.ylabel("Sales")

plt.title("Product Sale Comparison")
plt.legend()
plt.grid(color= "gray", linestyle=":")

plt.show()
```

```python
# To show data horizontally --> plt.barh
# bar chart --> plt.bar(x, height, color = "cname", width = value, label = "label name")

product = ["a","b","c","d"]
sales = [1000,1500,800,1200]

plt.barh(product, sales, color ="orange", label = "Sales 2025")

plt.xlabel("Products")
plt.ylabel("Sales")

plt.title("Product Sale Comparison")
plt.legend()
plt.grid(color= "gray", linestyle=":")

plt.show()
```

##### Pie chart

```python
# plt.pie(values, 
#           labels = label_list,
#           color = color_list, 
#           autopct = "%1.1f%%" )


regions = ["North", "West", "East", "South" ]
revenue = [3000,2000,1500,1000]

plt.pie(
        revenue,
        labels = regions, 
        autopct = "%1.1f%%", 
        colors = ["gold","skyblue", "lightgreen", "coral"] 
        )

plt.title("Revenue Contribution by Region")

plt.show()
```

#### Histogram

```python
# plt.hist(data, bins = number_of_bins, color = "clr_name", edgecolor = "black")

scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(
        scores,
         bins = 5,
         color = "coral",
         edgecolor = "black"
         )

plt.xlabel("Score Range")
plt.ylabel("Number od students")
plt.title("Score Distribution of Students")
plt.show()
```
