```python
import matplotlib.pyplot as plt 
```

```python
# plt.scatter(x,y, color, label, marker)

study_hours = [1,2,3,4,5,6,7,8]
exam_score = [50,55,60,65,70,75,80,85]

# scatter plot
plt.scatter(
    study_hours,
    exam_score,
    color = "green",
    marker = "o",
    label = "Student Data"
)

plt.xlabel("Study Hours")
plt.ylabel("Exam Score")

plt.title("study time and exam score relation")

plt.grid()
plt.show()
```

```python
# two plots in one canvas
plt.scatter(
    [1,2,3],
    [50,55,68],
    color = "green",
    marker = "o",
    label = "Class A"
)
plt.scatter(
    [1,2,3],
    [45,50,55],
    color = "orange",
    marker = "o",
    label = "Class B"
)

plt.xlabel("Study Hours")
plt.ylabel("Exam Score")

plt.title("Comparison of two classes") 

plt.legend()
plt.grid()
plt.show()
```
